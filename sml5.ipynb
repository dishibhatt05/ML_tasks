```
# remoivng user id
filter_data=df.iloc[0:,2:].values
```

```
# now apply training and testing
features=filter_data[0:,0:2]
```

```
# label finding
label=filter_data[0:,2]
```

```
# training and testing
from sklearn.model_selection import train_test_split
```

```
X,x,Y,y=train_test_split(features,label,test_size=0.2)
```

```
#now apply features scalling
from sklearn.preprocessing import StandardScaler
```

```
sc=StandardScaler()
```

```
# apply scalling data in training data only
X=sc.fit_transform(X)
```

```
# testing data scalling
x=sc.transform(x)
```

```
# knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```
# random fortest call eith number of decision tree
rclf=RandomForestClassifier(n_estimators=20)
```

```
# training
trained=rclf.fit(X,Y)
```
